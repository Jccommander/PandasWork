### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "CSV Files/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
print("Total Number of Players = " + str(len(purchase_data["SN"].value_counts())))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Perform calculations to obtain desired data
    #Average the values in the dataframe column "Price"
avg_price = round(purchase_data["Price"].mean(),2)
    #Sum the values in "Price"
total_price = purchase_data["Price"].sum()
    #Find the total number of unique items using a value count on "Item ID"
unique_items = len(purchase_data["Item ID"].value_counts())
    #Get the total number of items in the dataframe
total_row = purchase_data["SN"].count()

#Take all variables from above and put them into their own dataframe
purchase_analysis = pd.DataFrame({"Number of Unique Items":[unique_items],"Average Price":[avg_price],"Number of Purchases"\
                                 :[total_row],"Total Revenue":[total_price]})

#Clean up the dataframe by using .map to round and add "$" to monetary values
purchase_analysis["Average Price"] = purchase_analysis["Average Price"].map("${:}".format)
purchase_analysis["Total Revenue"] = purchase_analysis["Total Revenue"].map("${:}".format)

#Display the dataframe
purchase_analysis
```

## Gender Demographics

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Create a groupby on the Gender column to group genders
gender_group = purchase_data.groupby(["Gender"])

#Calculate the percentages for each gender out of the total
gender_percent = gender_group["SN"].count() / total_row * 100

#Create a new Data Frame to summarize the data
gender_table = gender_group.count()

#Add the percentage data to the newly created data frame
gender_table["Percentage"] = gender_percent

#Specify the columns that should be featured in the data frame
gender_complete = gender_table[["SN","Percentage"]]

#Rename the columns in order to give the data frame a cleaner appearance
gender_complete = gender_complete.rename(columns={"SN":"Total Count","Percentage":"Percentage of Players"})

#Use .map formatting to display the percentages
gender_complete["Percentage of Players"] = gender_complete["Percentage of Players"].map("{:.2f}%".format)

#Display the data frame
gender_complete
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Run calculations to determine the total purchases, average price and normalized totals for each gender
    #Use the gender groupby to sum the price values across each gender
total_price_gender = gender_group["Price"].sum()

    #Use the groupby to average the price values
avg_price_gender = gender_group["Price"].mean()
    
    #Take the total sum of prices for each gender and divide it by the total count of each gender to get the normalized total
normalized_gender = total_price_gender / purchase_data["Gender"].value_counts()

#Create a dataframe to house the above values
gender_temp = gender_group.count()

#Add each value to the dataframe under new columns
gender_temp["Average Purchase Price"] = avg_price_gender
gender_temp["Total Purchase Value"] = total_price_gender
gender_temp["Normalized Totals"] = normalized_gender

#Create a final dataframe that only holds the columns of interest
gender_analysis = gender_temp[["SN","Average Purchase Price","Total Purchase Value","Normalized Totals"]]

#Rename the columns and add .map formatting to display the resulting values as dollar amounts
gender_analysis = gender_analysis.rename(columns={"SN":"Purchase Count"})
gender_analysis["Average Purchase Price"] = gender_analysis["Average Purchase Price"].map("${:.2f}".format)
gender_analysis["Total Purchase Value"] = gender_analysis["Total Purchase Value"].map("${:.2f}".format)
gender_analysis["Normalized Totals"] = gender_analysis["Normalized Totals"].map("${:.2f}".format)

#Display the finished dataframe
gender_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Create a new dataframe to be edited
age_df = purchase_data

#Add a column to the new dataframe that contains the binned information from the "Age" column
age_df["Total Count"] = pd.cut(purchase_data["Age"],age_bins,labels=group_names)

#Greate a groupby off of the newly created bin column
age_group = age_df.groupby(["Total Count"])

#Use the groupby to create a percentage for each age group out of the total number of items
age_percent = age_group["SN"].count() / total_row * 100

#Create a new dataframe to be used to display the information
age_demograph = age_group.count()

#Add the percentage to the display data frame
age_demograph["Percent of Players"] = age_percent

#Specify the columns of interest in the dataframe and call them
age_demograph = age_demograph[["SN","Percent of Players"]]

#Reformat the dataframe to clean it up, using a .map to display the values as a percent
age_demograph = age_demograph.rename(columns={"SN":"Total Count"})
age_demograph["Percent of Players"] = age_demograph["Percent of Players"].map("{:.2f}%".format)

#Display the data frame
age_demograph
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create a groupby based on the "SN" column
sn_group = purchase_data.groupby(["SN"])

#Sum and Average the "Price" column using the groupby
total_price_sn = sn_group["Price"].sum()
avg_price_sn = sn_group["Price"].mean()

#Create a dataframe to house the above values
sn_temp = sn_group.count()

#Add the values to the dataframe
sn_temp["Average Purchase Price"] = avg_price_sn
sn_temp["Total Purchase Value"] = total_price_sn

#Specify the columns of interest and rename columns to clean
sn_analysis = sn_temp[["Purchase ID","Average Purchase Price","Total Purchase Value"]]
sn_analysis = sn_analysis.rename(columns={"Purchase ID":"Purchase Count"})

#Sort the values in the dataframe in decending order on the "Total Purchase Value" column
sn_analysis_sort = sn_analysis.sort_values("Total Purchase Value", ascending=False)

#Use .map formating to display the values as dollar amounts
sn_analysis_sort["Average Purchase Price"] = sn_analysis_sort["Average Purchase Price"].map("${:.2f}".format)
sn_analysis_sort["Total Purchase Value"] = sn_analysis_sort["Total Purchase Value"].map("${:.2f}".format)

#Display the dataframe
sn_analysis_sort.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create a dataframe that contains the columns of interest
items_df = purchase_data[["Item ID","Item Name","Price"]]

#Create a groupby on both the Item ID & Item Name columns
items_group = items_df.groupby(["Item ID","Item Name"])

#Count the total number of items in the dataset and save it to a variable
item_count = items_group.count()

#Sum the "Price" column and save the values to a variable
item_total = items_group.sum()

#Get the individual item prices by summing the "Price" column and then dividing it by the count of each groupby item
item_price = items_group["Price"].sum() / items_group["Price"].count()

#Create a new dataframe to house the above values
item_temp = items_group.count()

#Add the values to the dataframe along with column names
item_temp["Purchase Count"] = item_count
item_temp["Total Purchase Value"] = item_total
item_temp["Item Price"] = item_price

#Specify the columns of interest in the dataframe
item_analysis = item_temp[["Purchase Count","Item Price","Total Purchase Value"]]

#Sort the dataframe on the "Purchase Count" column in descending order
item_analysis_sort = item_analysis.sort_values("Purchase Count",ascending=False)

#Use .map to format the cells to display the values as monetary amounts
item_analysis_sort["Item Price"] = item_analysis_sort["Item Price"].map("${:.2f}".format)
item_analysis_sort["Total Purchase Value"] = item_analysis_sort["Total Purchase Value"].map("${:.2f}".format)

#Display the dataframe
item_analysis_sort.head()
```
